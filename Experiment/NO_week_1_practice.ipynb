```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-10, 10, 1000)
y = x**2

plt.plot(x, y, label = "y = x^2")
plt.legend()
plt.grid(True)
plt.show()
```

```python
x = np.linspace(-10, 10, 1000)
y = x**3

plt.plot(x, y, label = "y = x^2")
plt.legend()
plt.grid(True)
plt.show()
```
