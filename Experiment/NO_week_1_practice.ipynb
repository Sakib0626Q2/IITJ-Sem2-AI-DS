```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-10, 10, 1000)
y = x**2

plt.plot(x, y, label = "y = x^2")
plt.legend()
plt.grid(True)
plt.show()
```

```python
x = np.linspace(-20, 20, 10000)
y = x**3

plt.plot(x, y, label = "y = x^5")
plt.legend()
plt.grid(True)
plt.show()
```

```python
x = np.linspace(0, 5, 100)
y = np.exp(x)
plt.plot(x, y, label = "y = e^x")

y2 = np.log(x)
plt.plot(x, y2, label = "y = ln(x)")

plt.legend()
plt.grid(True)
plt.show()
```

```python
x = np.linspace(0, 50, 100)
y = np.log(x)
plt.plot(x, y, label = "y = ln (x)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
x = np.linspace(0, 2*np.pi, 1000)
y = np.sin(x)
plt.plot(x, y, label = "y = sin (x)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
x = np.linspace(0, 2*np.pi, 1000)
y = np.cos(x)
plt.plot(x, y, label = "y = cos (x)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
x = np.linspace(0, 2*np.pi, 1000)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x, y1, label = "y1 = sin (x) ")
plt.plot(x, y2, label = "y2 = cos (x)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
x = np.linspace(0, 2*np.pi, 1000)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(y1, y2, label = "y = sin (x) against cos (x)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
x = np.linspace(-2*np.pi, 2*np.pi, 10000)
y = np.tan(x)

# Remove extreme values near asymptotes
y[np.abs(y) > 10] = np.nan

plt.plot(x, y, label="y = tan(x)")
plt.ylim(-10, 10)
plt.legend()
plt.grid(True)
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# 1. Create an empty undirected graph
G = nx.Graph()

# 2. Add nodes
G.add_nodes_from([1, 2, 3, 4, 5])

# 3. Add edges
G.add_edges_from([(1, 2), (1, 3), (2, 4), (3, 5), (4, 5)])

# 4. Analyze the graph (e.g., find the shortest path)
shortest_path = nx.shortest_path(G, source=1, target=5)
print(f"Shortest path from 1 to 5: {shortest_path}") # Output: [1, 3, 5] or [1, 2, 4, 5]

# 5. Draw the graph
nx.draw(G, with_labels=True, node_color='skyblue', edge_color='gray')
plt.show()
```
