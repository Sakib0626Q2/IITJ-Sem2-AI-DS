## Assignment 1

```python
temperature = np.array([20, 22, 24, 26, 28, 30, 32, 34])
sales = np.array([80, 85, 88, 95, 100, 105, 110, 120])


plt.scatter(temperature, sales)
plt.xlabel("Temperature (Â°C)")
plt.ylabel("Ice Cream Sales")
plt.title("Temperature vs Ice Cream Sales")


x_line = np.linspace(20, 34, 100)
coeff = np.polyfit(temperature, sales, 1)
y_line = coeff[0] * x_line + coeff[1]
plt.plot(x_line, y_line, color='red', label='Model Line')
plt.legend()
plt.show()

# Add a line equation y=m*x+c to the below plot
```
