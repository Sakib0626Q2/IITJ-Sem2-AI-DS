```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 50)
y_scatter = 2 * x + np.random.normal(0, 2, 50)
y_line1 = 2*x + 1
y_line2 = 1.5*x + 3

plt.figure(figsize = (8,6))

plt.scatter(x, y_scatter, color = '#ff6B6B',
            alpha = 0.6, s= 40, label = 'Data Points')
plt.plot(x, y_line1, linestyle = '--', color = "#77c6df")
plt.plot(x, y_line2, linestyle = '-.', color = "#7bdf77")

plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return (x-3)**2 + 1

def df_dx(x):
    return 2 * (x-3)

x_start = 0.5
learning_rate = 0.1  # Set learning rate
iterations = 20  # Set the number of iterations

x = x_start
x_history = [x]
y_history = [f(x)]

# Initialize the plot
plt.figure(figsize=(8, 6))

# Create x values for plotting the function curve
x_plot = np.linspace(-1, 6, 100)
y_plot = f(x_plot)
plt.plot(x_plot, y_plot, label=r"$f(x) = (x-3)^2 + 1$", color='blue')

for i in range(iterations):
    gradient = df_dx(x)
    x = x - learning_rate * gradient  # Update x using the gradient
    x_history.append(x)
    y_history.append(f(x))

    print(f"step {i+1}: x = {x:.3f}, f(x) = {f(x):.3f}")

    # Plot current position of x
    plt.scatter(x, f(x), color='red', zorder=5)
    plt.plot(x_history, y_history, color='green', linestyle='--', alpha=0.5)

# Customize plot appearance
plt.title("Gradient Descent on $f(x) = (x-3)^2 + 1$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```
